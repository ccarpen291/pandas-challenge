```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python

#This will make the variables that I want to store in the data frame
#Print the results for good measure to make sure that everything is running correctly
columns = ["Top Players"]
df = pd.DataFrame(columns)

uniquecount = purchase_data["SN"].unique()
uniquecountvar = len(uniquecount)
print(f"Unique Player Number is {uniquecountvar}")

totalrevenue = purchase_data["Price"].sum()
print(f"Total Revenue is {totalrevenue}")

uniqueitemscount = purchase_data["Item ID"].unique()
uniqueitemsvar = len(uniqueitemscount)
print(f"Unique Items are {uniqueitemsvar}")


averagePrice = round(purchase_data["Price"].mean(),2)
print(f"Average Price is {round(averagePrice,2)}")


maxPrice = round(purchase_data["Price"].max(),2)
maxPrice

avgAge = round(purchase_data["Age"].mean(),2)
avgAge
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [-1,9,14,19,24,29,34,39,46]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data['Binned Ages'] = pd.cut(purchase_data['Age'], bins= bins, labels=group_names)




mean_df = purchase_data.groupby("Binned Ages")
age_count = mean_df["SN"].nunique()
age_percent = age_count/uniquecountvar
age_percent

age_count_percent = {"Number of Players":age_count,"Age Percentage":age_percent}

age_df_count_percent = pd.DataFrame(age_count_percent)
age_df_count_percent['Age Percentage'] = pd.Series(["{0:.2f}%".format(val*100) for val in age_df_count_percent['Age Percentage']], index = age_df_count_percent.index)
age_df_count_percent


#df['var3'] = pd.Series(["{0:.2f}%".format(val * 100) for val in df['var3']], index = df.index)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting

```python
#Purchase Count
#Average Purchase Price
#Total Purcahse Value


#Average Data Frame
price_average_series_df = purchase_data.groupby("SN").mean()["Price"]

#Count
count_df = purchase_data.groupby("SN").count()["Price"]

#Sum
sum_df = purchase_data.groupby("SN").sum()["Price"]

testdf = pd.DataFrame({"Avg Price ($USD)":price_average_series_df,"Purchase Count":count_df,"Total Purchase Value($USD)":sum_df})

testdf = testdf.sort_values("Purchase Count", ascending = False)
topfive = testdf.head(5)
topfive




#family_and_generosity = happiness_df.sort_values(
#    ["Family", "Generosity"], ascending=False)
#family_and_generosity.head()
```
